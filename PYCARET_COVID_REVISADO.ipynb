```python
import pandas as pd
from pycaret.classification import *

df = pd.read_csv('99b42b09-95af-47de-8411-ab99c380c3ef.csv')

df['descricao_dose'] = df['descricao_dose'].astype(int)

df = pd.get_dummies(df, columns=['sexo', 'raca_cor'], drop_first=True)
target = 'descricao_dose'

clf = setup(df, target=target)

best_model = compare_models()

final_model = finalize_model(best_model)

evaluate_model(final_model)

save_model(final_model, 'classificacao_covid')

loaded_model = load_model('classificacao_covid')

predictions = predict_model(loaded_model, data=df)

print(predictions)
```
